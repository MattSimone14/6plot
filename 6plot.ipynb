```python
#Submit a link to a GitHub repo where you have made 6 different plots using at least 2 plotting packages.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load the data
df = pd.read_csv("imdb_top_250.csv")

df_sample = df.head(20)

pandas_plot = df_sample.plot(kind='bar', x='Title', y='Year');
pandas_plot.set_xlabel('title');
pandas_plot.set_ylabel('Year');
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df_sample['Rating'], bins=10, kde=True, color='blue')
plt.xlabel("IMDb Rating")
plt.ylabel("Frequency")
plt.title("Distribution of IMDb Ratings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df_sample['Runtime'], df_sample['Rating'], color='red', alpha=0.6)
plt.xlabel("Runtime (minutes)")
plt.ylabel("IMDb Rating")
plt.title("Movie Runtime vs. IMDb Rating")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df_sorted = df_sample.sort_values(by="Year")
plt.plot(df_sorted["Year"], df_sorted["Rating"], linestyle='-', marker='o', color='purple')
plt.xlabel("Release Year")
plt.ylabel("IMDb Rating")
plt.title("IMDb Ratings Trend Over Time")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df_sample[['Rank', 'Rating', 'Runtime', 'Year']].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between IMDb Movie Features")
plt.show()
```

```python
sns.displot(df, x='Rating', bins=30, kind='hist', color='Purple', kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```
